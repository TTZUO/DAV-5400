```python
# import numpy,pandas libraries
import numpy as np
import pandas as pd
```

### 1&2. transform untidy data to tidy data
Create a csv file and upload to github, use pd.read_csv from github

```python
data = pd.read_csv("https://raw.githubusercontent.com/TTZUO/DAV-5400/master/untidy_data.csv")
# print data
data
```

Use dropna to drop null row

```python
data = data.dropna(how='all')
#print data
data
```

use rename to Rename first two columns as Airlines, Status

```python
data = data.rename(columns={"Unnamed: 0": "Airlines", "Unnamed: 1": "Status"})
# print data
data
```

Use fillna function to fill null value

```python
data['Airlines'].fillna(method='ffill', inplace = True)
# print data
data
```

Use melt function to group and rearrange data and print the tidy data

```python
data = data.melt(id_vars = ['Airlines', 'Status'], 
                 value_vars = ['Los Angeles', 'Phoenix', 'San Diego','San Francisco','Seattle'],
                var_name='Citys', value_name='Numbers')
# print data
data
```

### 3. For each city, which airline had the best on time performance?
Store on time data to dataframe on_time, use groupby function to group on_time's Airlines and citys and store in dataframe a, use groupby function to group data's Airlines and citys and store in new column a['total'], calcuate the on time rate and store in new column a['on_time_rate']

#### Answer: Airlines ALASK has best on time performance in each city.

```python
#Store on time data to dataframe on_time,
on_time = data[data['Status'] == 'on time']
#use groupby function to group on_time's Airlines and citys and store in dataframe a, 
a = on_time.groupby(['Airlines','Citys']).sum()
#use groupby function to group data's Airlines and citys and store in new column a['total']
a['total']=data.groupby(['Airlines','Citys']).sum()
#calcuate the on time rate and store in new column a['on_time_rate']
a['on_time_rate'] = a['Numbers']/a['total']
#print result
a
```
